```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
# Carregar microdados do enem
df = pd.read_csv('MICRODADOS_ENEM_2018.csv', encoding='Latin1', sep=';')
```

```python
#selecionar as provas apenas do estado do Rio Grande do Sul

rs = df[df['SG_UF_PROVA'] == 'RS']

rs.to_csv('rs.csv', index=False)

#rs = pd.read_csv('rs.csv')
```

```python
#selecionar apenas os estudantes que estiveram presentes nas 4 provas

rs2 = rs[(rs['TP_PRESENCA_CN'] == 1) & (rs['TP_PRESENCA_CH'] == 1) & (rs['TP_PRESENCA_LC'] == 1) & (rs['TP_PRESENCA_MT'] == 1)]
```

```python
rs2.head()
```

```python
#filtrar pelas colunas do questionário e outras de possível relevância
questoes = ['Q0' + str(x).zfill(2) for x in range(1,28)]
colunas = ['NU_IDADE', 'TP_SEXO', 'TP_COR_RACA', 'TP_ESTADO_CIVIL', 'TP_ST_CONCLUSAO', 'TP_ESCOLA', 'CO_ESCOLA', 
           'NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT', 'NU_NOTA_REDACAO']
colunas = colunas + questoes

rs3 = rs2.loc[:, colunas]
rs3.head()
```

```python
#salvar em csv
rs3.to_csv('paraModelo.csv', index = False)
```
